# Cleaning the data

```python
def snake_case(s):
    s = re.sub(r'\W+', '_', s)
    s=re.sub('(.)([A-Z][a-z]+)', r'\1_\2', s)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s).strip('_').lower()
```

```python
education.columns=[snake_case(col) for col in education.columns]
```

```python
condemns_per_region.columns=[snake_case(col) for col in condemns_per_region.columns]
```

```python
# put the names in english
education.columns = ["gender","regions","education","period","percentage_total"]
condemns_per_region.columns = ["regions","age_group","year","percentage_total"]
```

```python
# in education data we have every year divided in 3 i will put them together
education.dtypes
```

```python
education["year"]=education["period"].str.extract(r"(\d{4})")
education["percentage_total"]=education["percentage_total"].str.replace(',', '.').astype(float)
```

```python
#getting only the t4 values which is the end of the education year
education = education[education["period"].str.endswith("T4")]
education
```

```python
education=education.drop(["period","gender","education"],axis=1)
```

```python
education
```

```python
education['regions'] = education['regions'].str.replace(r'^\d+\s', '', regex=True)
education
```

```python
# now with the other table  condemns_per_region
condemns_per_region
```

```python
condemns_per_region=condemns_per_region[condemns_per_region["regions"] != "Total Nacional"].copy()
condemns_per_region["region"]=condemns_per_region["regions"]
condemns_per_region=condemns_per_region.drop("regions",axis=1)
condemns_per_region
```

```python
condemns_per_region=condemns_per_region.drop("age_group",axis=1)
```

```python
condemns_per_region.columns=["year","total_condemns_percentage","regions"]
condemns_per_region
```

```python
condemns_per_region['year'] = condemns_per_region['year'].astype(str)
condemns_per_region['regions'] = condemns_per_region['regions'].astype(str)

education['year'] = education['year'].astype(str)
education['regions'] = education['regions'].astype(str)
```

```python
df_combined = pd.merge(condemns_per_region, education, on=['regions', 'year'], how='outer')

df_combined.columns = ["year","total_condemns","regions","percentage_achivied_education"]
df_combined
```

### Try a chi test 

```python
# first convert educational attainment percentage into a categorical data ,like for example high and low
threshold = df_combined["percentage_achivied_education"].median()
df_combined ["education_level"] = pd.cut(df_combined["percentage_achivied_education"],bins = [0,threshold,100],labels=["Low","High"])

#do the same with condemns
condemns_threshold = df_combined["total_condemns"].median()
df_combined["condemns_level"]=pd.cut(df_combined["total_condemns"],bins=[0,condemns_threshold, df_combined ["total_condemns"].max()],labels=["Low","High"])

```

```python
contingency_table=pd.crosstab(df_combined["education_level"],df_combined["condemns_level"])
contingency_table
```

```python
# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}")
print(f"p-value: {p}")
```

## Conclusions and insights
* The findings from the correlation and chi square test suggest that there is no evidence of a linear relationship or significant association between educational attainment and the number of condemns in the data provided.
* So there is no statistical evidence from the data we have to support the hypothesis that lower levels of educational attainment are associated with higher number of condemns

```python
# Create a copy of the df_combined to ensure we are not working on a slice
df_combined = df_combined.copy()

# Convert year column to integers in the copied df_combined
df_combined['year'] = df_combined['year'].astype(int)

# Create mappings for the years 2015 and 2019 from df_combined_elections
politics_2015_mapping = df_combined_elections[df_combined_elections['years'] == 2015].set_index('regions')[['politic', 'politic_wing']].to_dict('index')
politics_2019_mapping = df_combined_elections[df_combined_elections['years'] == 2019].set_index('regions')[['politic', 'politic_wing']].to_dict('index')

# Initialize new columns for politic and politic_wing in df_combined using .loc for safe assignment
df_combined.loc[:, 'politic'] = None
df_combined.loc[:, 'politic_wing'] = None

# Iterate over the rows in df_combined and assign the correct values from mappings
for index, row in df_combined.iterrows():
    region = row['regions']
    year = row['year']
    # Apply the 2015 politic info if year is between 2015 and 2018
    if year <= 2018:
        politic_info = politics_2015_mapping.get(region)
        if politic_info:
            df_combined.loc[index, 'politic'] = politic_info['politic']
            df_combined.loc[index, 'politic_wing'] = politic_info['politic_wing']
    # Apply the 2019 politic info if year is between 2019 and 2022
    elif year <= 2022:
        politic_info = politics_2019_mapping.get(region)
        if politic_info:
            df_combined.loc[index, 'politic'] = politic_info['politic']
            df_combined.loc[index, 'politic_wing'] = politic_info['politic_wing']
```

```python
# group the data 
grouped_df = df_combined.groupby(['year', 'regions', 'politic_wing'], as_index=False)['total_condemns'].sum()
grouped_df
```

```python
# Calculate Condemn Rates per 100,000 people
# Assuming a constant population of 100,000 for each region for simplification

# Adding a new column for condemn rate per 100,000 people
# Here we're assuming that the total_condemns column already represents the total number of condemns
df_combined['condemn_rate_per_100k'] = (df_combined['total_condemns'] / 100) * 100000
```

```python
# Step 2: Calculate Year-Over-Year Change
# This requires shifting the condemn rate per 100,000 people by one year to calculate the difference
df_combined.sort_values(by=['regions', 'year'], inplace=True)
df_combined['yearly_change'] = df_combined.groupby('regions')['condemn_rate_per_100k'].pct_change() * 100
```

```python
# We will proceed with the above steps and then display the dataframe
df_combined[['year', 'regions', 'politic_wing', 'total_condemns', 'condemn_rate_per_100k', 'yearly_change']].head()
df_combined
```

```python
# Calculate the average condemn rate for each political wing
average_condemn_rate = df_combined.groupby('politic_wing')['condemn_rate_per_100k'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=average_condemn_rate, x='politic_wing', y='condemn_rate_per_100k')

plt.title('Average Condemn Rate per 100,000 People by Political Wing')
plt.xlabel('Political Wing')
plt.ylabel('Average Condemn Rate per 100,000 People')

plt.show()
```

* Null Hypothesis (0): There is no significant difference in condemn rates per 100,000 people between the 'Derechas' and 'Izquierdas' political wings. Any observed difference is due to random chance.
* Alternative Hypothesis (1): There is a significant difference in condemn rates per 100,000 people between the 'Derechas' and 'Izquierdas' political wings.

```python
# Filter the DataFrame by political wing
condemns_derechas = df_combined[df_combined['politic_wing'] == 'Derechas']['condemn_rate_per_100k']
condemns_izquierdas = df_combined[df_combined['politic_wing'] == 'Izquierdas']['condemn_rate_per_100k']

# Perform a t-test to determine if there's a significant difference in condemn rates between the political wings
t_stat, p_value = stats.ttest_ind(condemns_derechas, condemns_izquierdas, equal_var=False, nan_policy='omit')

t_stat, p_value
```

```python
# based on the p_value being less thath 0.05 we would reject the null hypothesis in favor of the alternative one.
```
